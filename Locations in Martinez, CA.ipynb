```python
import arcgis
from arcgis import *

# GIS Module
gis = GIS()

# Describing Map  Features 
map_martinez = gis.map('Martinez, CA')
map_martinez.basemap = 'satellite'
map_martinez.zoom = 13

# Locations
ran_aqua = geocode('100 Buckley St, Martinez, CA 94553')[0]
mar_museum = geocode('1005 Escobar St, Martinez, CA 94553')[0]
ran_park = geocode('Buckley St, Martinez, CA 94553')[0]
home = geocode('238 Arana Dr, Martinez, CA 94553')[0]

# Creating a List from Locations
mart_locations = ('ran_aqua', 'mar_museum', 'ran_park', 'home')

# Pop-up from Location 1
p_ran_aqua = {
    'title':'Rankin Aquatic Center',
    'content': ran_aqua['address']
}

# Symbol from Variable 1
pushpin_y={
"angle":0,
"xoffset":2,
"yoffset":8,
"type":"esriPMS",
"url":"http://static.arcgis.com/images/Symbols/Basic/GoldShinyPin.png",
"contentType":"image/png",
"width":24,
"height":24
}

# Pop-up from Location 2
p_mar_museum = {
    'title':'Martinez Museum',
    'content': mar_museum['address']
}

# Symbol from Variable 2
pushpin_r = {
"angle":0,
"xoffset":2,
"yoffset":8,
"type":"esriPMS","url":"http://static.arcgis.com/images/Symbols/Basic/RedShinyPin.png",
"contentType":"image/png",
"width":24,
"height":24
}

# Pop-up from Location 3
p_ran_park = {
    'title':'Rankin Park',
    'content': ran_park['address']
}

# Symbol from Variable 3
pushpin_g = {
"angle":0,
"xoffset":2,
"yoffset":8,
"type":"esriPMS","url":"http://static.arcgis.com/images/Symbols/Basic/LightGreenShinyPin.png",
"contentType":"image/png",
"width":24,
"height":24
}

# Pop-up from home
p_home = {
    'title':'Tibisay\'s Home',
    'content': home['address']
}

# Symbol from home
home_symbol = {
"angle":0,
"xoffset":0,
"yoffset":0,
"type":"esriPMS",
"url":"http://static.arcgis.com/images/Symbols/PeoplePlaces/Housing.png",
"contentType":"image/png"
,"width":24,
"height":24
}

# Creating a For loop for items in list of locations
for location in mart_locations:
    if location == 'ran_aqua':
        map_martinez.draw(ran_aqua['location'], p_ran_aqua, pushpin_y)
    elif location == 'mar_museum':
        map_martinez.draw(mar_museum['location'], p_mar_museum, pushpin_r)
    elif location == 'ran_park':
        map_martinez.draw(ran_park['location'], p_ran_park, pushpin_g)
    else:
        map_martinez.draw(home['location'], p_home, home_symbol)
        
# Mapping Locations in Martinez


map_martinez

# map_martinez.take_screenshot()
```
